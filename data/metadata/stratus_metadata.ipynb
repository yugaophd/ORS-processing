```python
import xarray as xr
import json
```

```python
ds = xr.open_dataset('OS_NTAS_2016_D_TS.nc')
```

## Export metadata to a JSON file

```python
metadata = {
    'dimensions': dict(ds.dims),
    'coordinates': {name: {'data': coord.values.tolist(), 'attrs': dict(coord.attrs)} for name, coord in ds.coords.items()},
    'variables': {name: {'dims': var.dims, 'data_type': str(var.dtype), 'attrs': dict(var.attrs)} for name, var in ds.data_vars.items()},
    'attributes': dict(ds.attrs)
}

# Export metadata to a JSON file
with open('OS_NTAS_2016_D_TS.json', 'w') as f:
    json.dump(metadata, f, indent=4)
```

## Additionally, export metadata to a TXT file with custom formatting

```python
with open('OS_NTAS_2016_D_TS.txt', 'w') as f_txt:
    f_txt.write("Dimensions and Sizes:\n")
    for dim, size in metadata['dimensions'].items():
        f_txt.write(f"- {dim}: {size}\n")
    
    f_txt.write("\nCoordinates:\n")
    for coord, details in metadata['coordinates'].items():
        f_txt.write(f"- {coord}: {details['data']} (Attributes: {details['attrs']})\n")
    
    f_txt.write("\nVariables:\n")
    for var, info in metadata['variables'].items():
        f_txt.write(f"- {var}: {info['dims']}, Type: {info['data_type']} (Attributes: {info['attrs']})\n")
    
    f_txt.write("\nGlobal Attributes:\n")
    for attr, value in metadata['attributes'].items():
        f_txt.write(f"- {attr}: {value}\n")
```
